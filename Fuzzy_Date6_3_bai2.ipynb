```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

#Sensor
sensor_left = ctrl.Antecedent(np.arange(0,1,0.1),'sensor_left') 
sensor_center = ctrl.Antecedent(np.arange(0,0.8,0.1),'sensor_center')
sensor_right = ctrl.Antecedent(np.arange(0,0.9,0.1),'sensor_right')  

#PWM
pwm_left = ctrl.Antecedent(np.arange(0,256,0.1),'pwm_left') 
pwm_right = ctrl.Antecedent(np.arange(0,256,0.1),'pwm_right') 

sensor_left['sidebyside'] = fuzz.trimf(sensor_left.universe,[0,0,0.4]) 
sensor_left['far'] = fuzz.trimf(sensor_left.universe,[0,0.4,0.9]) 
sensor_left['sofar'] = fuzz.trimf(sensor_left.universe,[0.4,0.9,0.9]) 

sensor_center['sidebyside'] = fuzz.trimf(sensor_center.universe,[0,0,0.3]) 
sensor_center['far'] = fuzz.trimf(sensor_center.universe,[0,0.3,0.7]) 
sensor_center['sofar'] = fuzz.trimf(sensor_center.universe,[0.3,0.7,0.7]) 

sensor_right['sidebyside'] = fuzz.trimf(sensor_right.universe,[0,0,0.4]) 
sensor_right['far'] = fuzz.trimf(sensor_right.universe,[0,0.4,0.8]) 
sensor_right['sofar'] = fuzz.trimf(sensor_right.universe,[0.4,0.8,0.8])

sensor_left.view()
sensor_center.view()
sensor_right.view()

pwm_left['veryslow'] = fuzz.trimf(pwm_left.universe,[0,0,35]) 
pwm_left['slow'] = fuzz.trimf(pwm_left.universe,[0,35,100]) 
pwm_left['aver'] = fuzz.trimf(pwm_left.universe,[35,100,200]) 
pwm_left['fast'] = fuzz.trimf(pwm_left.universe,[100,200,255]) 
pwm_left['veryfast'] = fuzz.trimf(pwm_left.universe,[200,255,255]) 


pwm_right['veryslow'] = fuzz.trimf(pwm_right.universe,[0,0,35]) 
pwm_right['slow'] = fuzz.trimf(pwm_right.universe,[0,35,100]) 
pwm_right['aver'] = fuzz.trimf(pwm_right.universe,[35,100,200]) 
pwm_right['fast'] = fuzz.trimf(pwm_right.universe,[100,200,255]) 
pwm_right['veryfast'] = fuzz.trimf(pwm_right.universe,[200,255,255]) 

pwm_right.view()
pwm_left.view()

```
