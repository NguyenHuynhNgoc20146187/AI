```python
from keras.models import load_model
model100 = load_model('/content/drive/MyDrive/Colab Notebooks/model_cifar100.h5')
```

```python
import matplotlib.pyplot as plt
from keras.utils import load_img, img_to_array
import numpy as np
img = load_img('/content/drive/MyDrive/Colab Notebooks/image.jpg',target_size = (32,32))
plt.imshow(img)
plt.show()
img= img_to_array(img)

img = img.reshape(1, 32*32*3)
img = img.astype('float32')
img = img/255
np.argmax(model100.predict(img), axis = -1)
```
